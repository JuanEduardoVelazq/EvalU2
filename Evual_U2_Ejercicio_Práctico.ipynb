1. Random Forest (Bosque Aleatorio)

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

mipath = "australian.dat"
data = pd.read_csv(mipath, sep=" ", header=None)
data.columns = ["A1","A2","A3","A4","A5","A6","A7","A8","A9","A10","A11","A12","A13","A14","class"]

X = data.iloc[:,:-1]
y = data[["class"]]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.6, random_state=45)
```

Ciclo for con el modelo base

```python
n_for = 10
accAD = list()

for k in range(n_for):
    modelo_rf = RandomForestClassifier()
    modelo_rf.fit(X_train, y_train)

    tmp = modelo_rf.score(X_test, y_test)
    accAD.append(tmp)
print("Árbol de Decisión con parámetros por default\nPromedio y desviación estándar: %.1f%% (%.1f)" % (100*np.mean(accAD), 100*np.std(accAD)))

pd.DataFrame(accAD).boxplot()
plt.show()
```

Ciclo for con el modelo base con hiper parámetros

```python
accAD2 = list()

for k in range(n_for):
    modelo_rf = RandomForestClassifier(n_estimators=100,
                                       criterion="gini",
                                       max_depth=None,
                                       min_samples_split=2,
                                       min_samples_leaf=1,
                                       max_features=12)
    modelo_rf.fit(X_train, y_train)

    tmp = modelo_rf.score(X_test, y_test)
    accAD2.append(tmp)
print("Árbol de Decisión con parámetros por default\nPromedio y desviación estándar: %.1f%% (%.1f)" % (100*np.mean(accAD2), 100*np.std(accAD2)))

pd.DataFrame(accAD2).boxplot()
plt.show()
```

Graficar los bloxplot, ambos que se muestren en la misma gráfica

```python
ddff = pd.DataFrame(list(zip(accAD, accAD2)), columns=["default","ajustado"])
ddff.plot(kind='box')
plt.title('Árbol de Decisión: exactitud (accuracy)')
plt.show()
```

2. Linear Support Vector Classifier (LinearSVC)

```python
from sklearn.svm import LinearSVC
```

```python
accAD = list()

for k in range(n_for):
    modelo_rf = LinearSVC()
    modelo_rf.fit(X_train, y_train)

    tmp = modelo_rf.score(X_test, y_test)
    accAD.append(tmp)
print("Árbol de Decisión con parámetros por default\nPromedio y desviación estándar: %.1f%% (%.1f)" % (100*np.mean(accAD), 100*np.std(accAD)))

pd.DataFrame(accAD).boxplot()
plt.show()
```

Ciclo for con el modelo con hiper parámetros

```python
accAD2 = list()

for k in range(n_for):
    modelo_rf = LinearSVC(C=0.1,
                          penalty="l1",
                          dual=False)
    modelo_rf.fit(X_train, y_train)

    tmp = modelo_rf.score(X_test, y_test)
    accAD2.append(tmp)
print("Árbol de Decisión con parámetros por default\nPromedio y desviación estándar: %.1f%% (%.1f)" % (100*np.mean(accAD2), 100*np.std(accAD2)))

pd.DataFrame(accAD2).boxplot()
plt.show()
```

Graficar los bloxplot, ambos que se muestren en la misma gráfica

```python
ddff = pd.DataFrame(list(zip(accAD, accAD2)), columns=["default","ajustado"])
ddff.plot(kind='box')
plt.title('Árbol de Decisión: exactitud (accuracy)')
plt.show()
```

¿Qué modelo arroja mejores scores?

Podemos ver que en el modelo Random Forest de manera base tenemos un 88.7% de promedio y de desviación estandar tenemos un 0.6
En el caso del modelo LinearSVC tenemos que de manera base tenemos resultados con un promedio de 75% con una desviación estandar de 10.1.

Ahora hablando de los resultados cuando utilizamos los hiper parámetros tenemos como resultado que el modelo Random Forest tiene un mayor promedio con 89.5% y con una desviación estandar de 0.5. Por otro lado tenemos que el LinearSVC con un 89.1% y con una desvición de 0.0.
